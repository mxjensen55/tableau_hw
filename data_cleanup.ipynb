```python
# Import dependencies
import csv
import glob
import pandas as pd
```

```python
#merge all csv files
citibike_files = glob.glob("data/*.csv") 
df = pd.concat((pd.read_csv(f, header = 0) for f in citibike_files))
```

```python
#verify the shape of the data
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
#change gender to string
df['gender'] = df['gender'].astype(str)
```

```python
#replace number values with genders
gender_df = df['gender']
df_male = gender_df.replace('1', 'Male')
df_female = df_male.replace('2', 'Female')
df_unknown = df_female.replace('0', 'Unknown')
df['gender'] = df_unknown
df.head()
```

```python
#get the age of users
df['Age'] = pd.Timestamp('now').year - df['birth year']
df.head()
```

```python
# Export merged/cleaned file as a CSV
df.to_csv("data/citibike_final.csv", index=False)
```
